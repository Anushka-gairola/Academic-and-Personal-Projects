```python
# Install pymongo package
import subprocess
import sys

def install(package):
    subprocess.check_call([sys.executable, "-m", "pip", "install", package])
    
install("pymongo")

```

```python
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient('mongodb://localhost:27017/')

# Check the databases
print(client.list_database_names())
```

```python
# Function to check if the database exists 
def check_database_exists(db_name):
    local_dblist = client.list_database_names()
    if db_name in local_dblist:
        print("The database ", db_name, " exists.")
    else:
        print("The database ", db_name, " does not exist.")

# Function to check if the collection exists
def check_collection_exists(argDBName, argCollName, local_mydb):
    local_collist = local_mydb.list_collection_names()
    if argCollName in local_collist:
        print("The collection ",  argCollName, "exists in database ", argDBName)
    else:
        print("The collection ", argCollName, " does not exist in database ", argDBName)
```

2. Creating Database 

```python
      

# Defining database and collection names
mydb = client["BEMM459_GroupP"]   # Our chosen Database name


print(type(mydb))

#Checking if database exists by calling function check_DatabaseExists
check_database_exists("BEMM459_GroupP")

print()
```
